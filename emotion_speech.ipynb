```python
path=[]
labels =[]
rout = '/content/sample_data/emotions'
for direc ,_,files in os.walk(rout):
  for file in files :
    path.append(os.path.join(direc,file))
    label = os.path.basename(direc)


    labels.append(label.lower())
print(" data loaded")
print("Number of files:", len(path))
print("Number of labels:", len(labels))

# Print first few entries in path and labels
print("\nSample paths:")
for p in path[:5]:  # Print the first 5 paths
    print(p)

print("\nSample labels:")
for l in labels[:5]:  # Print the first 5 labels
    print(l)
```

```python
def extract_mfcc(fname):
    y,sr = librosa.load(fname, duration=5, offset = 0.5)
    return np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc = 40).T, axis =0)
```

```python
new_var = df[ ' the speech' ].iloc[-1]
extract_mfcc(new_var)
```

```python
X_mfcc = df[ ' the speech' ].apply(lambda x: extract_mfcc(x))
X_mfcc
X = [x for x in X_mfcc]
X =  np.array(X)
X.shape
```

```python
X = np.expand_dims(X, -1)
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
enc =  OneHotEncoder()
y = enc.fit_transform(df[['emotion']])

y=y.toarray()
y.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

model = Sequential([
    LSTM(123, return_sequences= False, input_shape= (40,1)),
    Dense(64, activation= 'relu'),
    Dropout(0.2),
    Dense(32, activation = 'relu'),
    Dropout(0.2),
    Dense(7, activation = 'softmax')

])

# Compiler le modle
model.compile(loss= 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.summary()
```

```python

model.summary()

```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Assurez-vous que vous n'avez pas redfini accidentellement une fonction ou une variable
# par un nom de chane de caractres ou autre type incompatible

# Activer le mode impatient pour le dbogage (si ncessaire)
tf.config.run_functions_eagerly(True)

# Diviser les donnes en ensembles d'entranement et de validation
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
print(type(model))
```

```python
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=16, shuffle=True)
```

```python
print(history.history)
```

```python
print(X.shape)
print(y.shape)
```

```python
epochs = list(range(100))
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.plot(epochs, acc, label='train accuracy')
plt.plot(epochs, val_acc, label='val accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs, loss, label='train loss')
plt.plot(epochs, val_loss, label='val loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
import librosa
import numpy as np
from tensorflow.keras.models import load_model
from sklearn.preprocessing import LabelEncoder

# Function to extract MFCC features from audio file
def extract_mfcc(fname, duration=5, offset=0.5, n_mfcc=40):
    y, sr = librosa.load(fname, duration=duration, offset=offset)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfccs.T, axis=0)  # Calculate mean across time to get a single feature vector

# Load the trained model
model = load_model('emotion_detection_model.h5')

# Example path to a test audio file
test_file = '/content/sample_data/emotions/DC_a01.wav'
test2= '/content/sample_data/emotions/OAF_death_fear.wav'

# Extract MFCC features and average over time
mfcc_features = extract_mfcc(test_file)
mfcc_features = extract_mfcc(test2)

# Add batch and channel dimensions to match model input
X_test = mfcc_features[np.newaxis, ..., np.newaxis]  # Add batch and channel dimensions

# Make predictions
predictions = model.predict(X_test)


encoder = LabelEncoder()
encoder.classes_ = np.array(['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise'])

predicted_label = encoder.inverse_transform(np.argmax(predictions, axis=1))

# Print predicted emotion
print("Predicted Emotion:", predicted_label[0])
Audio(test_file)
Audio(test2)
```
